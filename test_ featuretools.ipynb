```python
import featuretools as ft
from woodwork.logical_types import Categorical, PostalCode

data = ft.demo.load_mock_customer()
transactions_df = (
    data["transactions"].merge(data["sessions"]).merge(data["customers"])
)
products_df = data["products"]

es = ft.EntitySet()

es = es.add_dataframe(
    dataframe_name="transactions",
    dataframe=transactions_df,
    index="transaction_id",
    time_index="transaction_time",
    logical_types={
        "product_id": Categorical,
        "zip_code": PostalCode,
    },
)
es = es.add_dataframe(
    dataframe_name="products", dataframe=products_df, index="product_id"
)
es = es.add_relationship(
    "products", "product_id", "transactions", "product_id"
)
```

```python
class Absolute(TransformPrimitive):
    name = "absolute"
    input_types = [ColumnSchema(semantic_tags={"numeric"})]
    return_type = ColumnSchema(semantic_tags={"numeric"})

    def get_function(self):
        def absolute(column):
            return abs(column)

        return absolute


class Maximum(AggregationPrimitive):
    name = "maximum"
    input_types = [ColumnSchema(semantic_tags={"numeric"})]
    return_type = ColumnSchema(semantic_tags={"numeric"})

    def get_function(self):
        def maximum(column):
            return max(column)

        return maximum
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from autofeat import AutoFeatClassifier
import pandas as pd
import numpy as np

cat_features = [
    "paperless_billing",
    "payment_method",
    "internet_service",
    "online_security",
    "online_backup",
    "device_protection",
    "tech_support",
    "streaming_tv",
    "streaming_movies",
    "gender",
    "senior_citizen",
    "partner",
    "dependents",
    "multiple_lines",
]
num_features = ["monthly_charges", "total_charges"]

features = cat_features + num_features
target = "target"
split_column = "begin_date"
test_size = 0.2

df = df.sort_values(by=[split_column])

X_train, X_test, y_train, y_test = train_test_split(
    df[features],
    df[target],
    test_size=test_size,
    shuffle=False,
)

# transformations = ["1/", "exp", "log", "abs", "sqrt", "^2", "^3", "1+", "1-", "sin", "cos", "exp-", "2^"]
transformations = ("1/", "log", "abs", "sqrt")

afc = AutoFeatClassifier(
    categorical_cols=cat_features,
    transformations=transformations,
    feateng_steps=1,
    n_jobs=-1,
)

X_train_features = afc.fit_transform(X_train, y_train)
X_test_features = afc.transform(X_test)

# обучи модель с учетом новых признаков
from catboost import CatBoostClassifier
from sklearn.metrics import roc_auc_score

model = CatBoostClassifier(
    auto_class_weights="Balanced",
    random_seed=42,
    verbose=0,
)
model.fit(X_train_features, y_train)

y_test_proba = model.predict_proba(X_test_features)[:, 1]
roc_auc = roc_auc_score(y_test, y_test_proba)
print(f"ROC-AUC на тестовой выборке: {roc_auc:.4f}")
```

```python
from catboost import CatBoostClassifier
from sklearn.metrics import roc_auc_score

model = CatBoostClassifier(
    auto_class_weights="Balanced",
    random_seed=42,
    verbose=0,
)
model.fit(X_train_features, y_train)

y_test_proba = model.predict_proba(X_test_features)[:, 1]
roc_auc = roc_auc_score(y_test, y_test_proba)
print(f"ROC-AUC на тестовой выборке: {roc_auc:.4f}")

# сохрани модель в MLflow
import mlflow
import mlflow.catboost
from mlflow.models import infer_signature

mlflow.set_tracking_uri(TRACKING_SERVER_URL)
mlflow.set_registry_uri(TRACKING_SERVER_URL)

experiment = mlflow.get_experiment_by_name(EXPERIMENT_NAME)
if experiment is None:
    experiment_id = mlflow.create_experiment(EXPERIMENT_NAME)
else:
    experiment_id = experiment.experiment_id

signature = infer_signature(
    X_train_features,
    model.predict(X_train_features),
)

with mlflow.start_run(
    run_name="autofeat_classifier_run_5",
    experiment_id=experiment_id,
) as run:
    mlflow.log_params(
        {
            "model": "CatBoostClassifier",
            "auto_class_weights": "Balanced",
            "random_seed": 42,
            "features_count": X_train_features.shape[1],
        }
    )
    mlflow.log_metric("test_roc_auc", roc_auc)

    model_info = mlflow.catboost.log_model(
        cb_model=model,
        artifact_path="catboost_model",
        registered_model_name="cb_model_on_autofeat_features",
        signature=signature,
        input_example=X_train_features.head(5),
    )

    print(f"MLflow run_id: {run.info.run_id}")
    print(f"Модель сохранена: {model_info.model_uri}")
```
